5.How many global scopes are there in a Python program? How many local scopes?

```python
'''
In a Python program, there is only one global scope, and it exists throughout the entire program's execution.
This global scope contains all the variables and functions defined at the top level of the program or in modules imported into the program.
the number of local scopes can vary depending on how many functions are defined in the program and how often they are called.
Each time a function is called, a new local scope is created for that function's execution. When the function finishes executing,
its local scope is destroyed, and the program returns to the previous scope.

Global Scope: There is only one global scope for the entire program.
Local Scopes: The number of local scopes can change dynamically depending on how many functions are called during program execution. Each function call creates a new local scope for that function's execution.

'''
```

9.How do you make a function variable refer to the global variable?

```python
'''
In this example, we have a global variable global_var with an initial value of 10.
Inside the function function_with_global_var(), we declare the variable global_var as global using the global keyword.
Any assignment made to global_var inside the function will modify the global variable's value.
Without the global keyword, if you just assign a value to global_var inside the function,
Python would create a new local variable within the function's scope instead of modifying the global variable.
Using global allows you to access and modify the global variable directly from within the function.
'''
global_var = 10  # This is the global variable

def function_with_global_var():
    global global_var  # Use the global keyword to access the global variable inside the function
    global_var = 20    # Now, this modifies the global variable instead of creating a new local variable
print("before function call",global_var)
function_with_global_var()
print("after function call",global_var)  # Output: 20

```
