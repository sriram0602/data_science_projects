Q.3. What are the standards and conventions followed for the nomenclature of variables in
python to improve code readability and maintainability?<br>
1. **Use descriptive names**: Choose meaningful names that describe the purpose or content of the variable. This helps others (including yourself in the future) understand the code without the need for excessive comments.

2. **Use lowercase letters**: Variable names should be in all lowercase letters, separated by underscores, if necessary. For example: `num_students`, `average_score`, `max_value`.

3. **Avoid single-character names**: Unless the variable represents an iteration variable (e.g., `i` for loops), avoid using single-character names like `x`, `y`, or `z`. Instead, use descriptive names even for short-lived variables.

4. **Be concise but not cryptic**: Keep variable names reasonably short while still being descriptive. Avoid using abbreviations that may be unclear to others.

5. **Use nouns for object names**: Variables typically represent objects, so using nouns is more appropriate. For instance, use `customer_name` rather than `get_name()`.

6. **Use underscores for readability**: To improve readability, separate words in variable names using underscores. This is often referred to as "snake_case."

7. **Constants in uppercase**: If you have constants, their names should be written in uppercase letters. If a constant name consists of multiple words, use underscores to separate them. For example: `MAX_SIZE`, `PI_VALUE`.

8. **Avoid using built-in names**: Don't use names of built-in functions or modules (e.g., `list`, `dict`, `file`, `str`, etc.) as variable names, as it can lead to unexpected behavior and confusion.

Q.9.Write a code to create the given structure using only for loop.

```python
n=5
lenj=2*n-1
for i in range(n):
  tabj=n-(i+1)
  for j in range(lenj):
    if j < tabj or j >lenj-tabj-1:
      print(end=' ')
    else:
      print('*' ,end=' ')
  print("\n")
```

Q.10. Write a code to create the given structure using while loop.

```python
n=5
lenj=2*n-1
for i in range(n,0,-1):
  tabj=n-i
  for j in range(lenj):
    if j < tabj or j >lenj-tabj-1:
      print(end=' ')
    else:
      print('|' ,end=' ')
  print("\n")
```
