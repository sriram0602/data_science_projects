10. How can you tell the difference between break and continue?

```python
#break- This keyword will break the entire loop and comes out of it
#continue- This keyword will skip the current iteration and moves to the next one

for i in range(1,5):
  # breaks the loop when i has value 3
    if i == 3:
        break
    print(i)

#continue
for i in range(1,10):
    #skips the iteration when i has value 5 and moves to next value
    if i == 5:
        continue
```

11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

```python
# range(10) -> prints the value from 0 to 9 excluding 10,by incrementing value as 1
print("range(10)")
for i in range(10):
  print(i,end=" ")
#range(0,10)->prints the values which has len 10 with starting number as 0 ,and ending as 9 excluding 9,with increment value as 1 by default
print("\nrange(0,10)")
for i in range(0,10):
  print(i,end=" ")
#range(0,10,1)->prints the values which has len 10 with starting number as 0 ,and ending as 9 excluding 9,with increment value as 1 which is given at last
print("\nrange(0,10,1)")
for i in range(0,10):
  print(i,end=" ")
```

12.Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent

```python
#for loop
print("using for loop")
for i in range(1,11):
  print(i,end=" ")
#while loop
print("\nusing while loop")
j=1
while j<=10:
   print(j,end=" ")
   j+=1

```
