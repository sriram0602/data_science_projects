5. If a dictionary is stored in spam, what is the difference between the expressions &#39;cat&#39; in spam and
&#39;cat&#39; in spam.keys()?

```python
# the both expression checks whether the key value cat is in the dict.
spam={"cat":"domestic","tiger":"wild"}

print("cat" in spam)
print("cat" in spam.keys())
```

6. If a dictionary is stored in spam, what is the difference between the expressions &#39;cat&#39; in spam and
&#39;cat&#39; in spam.values()?

```python
spam={"cat":"domestic","tiger":"wild"}

#the expression cat in spam checks whether the key cat is in dict and returns true
#since it is in list
print("cat" in spam)
# the expression the cat in spam.values checkws whether the value cat is in the dict and returns false
#since it is not in list
print("cat" in spam.values())
```

7. What is a shortcut for the following code?
